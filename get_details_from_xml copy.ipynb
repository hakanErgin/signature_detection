```python
import xml.etree.ElementTree as ET
import glob
```

```python
class Signatures:
    signatures = []

    def __init__(self):
        self.signatures = [self.get_signatures(file) for file in glob.glob("./assets/train_xml/*.xml")]

    def get_signatures(self, file):
        root = ET.parse(file).getroot()
        if len(root.findall(".//*[@gedi_type='DLSignature']")) > 0:
            # print(file)
            return 1
        else :
            return 0
```
